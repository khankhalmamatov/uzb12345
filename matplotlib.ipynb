```python
# BAR CHART
data['Gender'].value_counts()
labels = np.array(["Erkaklar", "Ayollar"])
counts = np.array([245, 255]) 
plt.bar(labels[0],counts[0],label = "erkaklar", color="skyblue")
plt.bar(labels[1], counts[1],label = "ayollar", color="pink")
plt.title("Ayollar va erkaklar soni")
plt.xlabel("Gender")
plt.ylabel("odamlar soni")
plt.legend(loc='upper left')
plt.grid(axis="y")
plt.show()
```

```python
# PIA CHART
gender = np.array(["Erkaklar", "Ayollar"])
values = np.array([245,255])
colors = ["skyblue","pink"]
plt.pie(values, labels=gender, autopct="%1.1f%%", colors=colors)
plt.show()
```
