```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

index_counts = df['Index'].value_counts().sort_index()

fig = plt.figure(figsize=(6,5))
ax = fig.add_subplot(111, projection='3d')

x = np.arange(len(index_counts))
y = np.zeros(len(index_counts))  # barlar y oâ€˜qi boâ€˜yicha boshlanishi
z = np.zeros(len(index_counts))  # z=oâ€˜qi boâ€˜yicha boshlanishi
dx = 0.4 * np.ones(len(index_counts))  # bar kengligi
dy = 0.4 * np.ones(len(index_counts))
dz = index_counts.values  # balandlik = odamlar soni

ax.bar3d(x, y, z, dx, dy, dz, color='skyblue', shade=True)

ax.set_xticks(x)
ax.set_xticklabels(index_counts.index)
ax.set_xlabel('Indexlar')
ax.set_ylabel('Y oâ€˜qi')
ax.set_zlabel('Odamlar soni')
ax.set_title("Har bir indexdagi holatlar soni (3D Bar Chart)")

plt.show()
```
