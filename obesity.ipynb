```python
# Faqat ayollarni ajratib olamiz
df_female = df[df['Gender'] == 'Female']

female_index_counts = df_female['Index'].value_counts().sort_index()

# PIE CHART â€“ ayollardagi ulush
plt.figure(figsize=(5,5))
plt.pie(female_index_counts.values,
        labels=female_index_counts.index,
        autopct='%1.1f%%',
        startangle=90)
plt.title("Ayollar kesimida indexlar ulushi (Pie chart)")
plt.show()

# BAR CHART â€“ ayollardagi miqdor
plt.figure(figsize=(5,4))
plt.bar(female_index_counts.index, female_index_counts.values, width=0.4)
plt.title("Ayollar kesimida indexlar miqdori (Bar chart)")
plt.xlabel("Indexlar")
plt.ylabel("Ayollar soni")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
