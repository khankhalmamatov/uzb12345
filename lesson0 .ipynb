```python
import pandas as pd
import numpy as np

# ===== DATA =====
df = pd.read_csv('Titanic dataset.csv')
df.drop(columns=['home.dest', 'name', 'cabin', 'ticket', 'body', 'embarked'], inplace=True)

# ================== 1-KLASS ==================
print("BIRINCHI KLASS BO'YICHA OMON QOLGANLAR")

x = (df['pclass'] == 1).sum()
y = ((df['pclass'] == 1) & (df['survived'] == 1)).sum()
z = (y / x * 100)
print(f"Jami odamlar soni: {x}")
print(f"Omon qolganlar foizi: {z:.2f}%")

birinchi_klass = df[df['pclass'] == 1]

print("Yosh bo'yicha qiymatlar")
print(f"O'rtacha yosh: {birinchi_klass['age'].mean():.2f}")
print(f"Minimal yosh: {birinchi_klass['age'].min():.2f}")
print(f"Maximal yosh: {birinchi_klass['age'].max():.2f}")

print("Yo'l haqi bo'yicha qiymatlar")
print(f"O'rtacha yo'l haqi: {birinchi_klass['fare'].mean():.2f} $")
print(f"Minimal yo'l haqi: {birinchi_klass['fare'].min():.2f} $")
print(f"Maximal yo'l haqi: {birinchi_klass['fare'].max():.2f} $")

print("\n" + "-"*50 + "\n")

# ================== 2-KLASS ==================
print("IKKINCHI KLASS BO'YICHA OMON QOLGANLAR")

x = (df['pclass'] == 2).sum()
y = ((df['pclass'] == 2) & (df['survived'] == 1)).sum()
z = (y / x * 100)
print(f"Jami odamlar soni: {x}")
print(f"Omon qolganlar foizi: {z:.2f}%")

ikkinchi_klass = df[df['pclass'] == 2]

print("Yosh bo'yicha qiymatlar")
print(f"O'rtacha yosh: {ikkinchi_klass['age'].mean():.2f}")
print(f"Minimal yosh: {ikkinchi_klass['age'].min():.2f}")
print(f"Maximal yosh: {ikkinchi_klass['age'].max():.2f}")

print("Yo'l haqi bo'yicha qiymatlar")
print(f"O'rtacha yo'l haqi: {ikkinchi_klass['fare'].mean():.2f} $")
print(f"Minimal yo'l haqi: {ikkinchi_klass['fare'].min():.2f} $")
print(f"Maximal yo'l haqi: {ikkinchi_klass['fare'].max():.2f} $")

print("\n" + "-"*50 + "\n")

# ================== 3-KLASS ==================
print("UCHINCHI KLASS BO'YICHA OMON QOLGANLAR")

x = (df['pclass'] == 3).sum()
y = ((df['pclass'] == 3) & (df['survived'] == 1)).sum()
z = (y / x * 100)
print(f"Jami odamlar soni: {x}")
print(f"Omon qolganlar foizi: {z:.2f}%")

uchinchi_klass = df[df['pclass'] == 3]

print("Yosh bo'yicha qiymatlar")
print(f"O'rtacha yosh: {uchinchi_klass['age'].mean():.2f}")
print(f"Minimal yosh: {uchinchi_klass['age'].min():.2f}")
print(f"Maximal yosh: {uchinchi_klass['age'].max():.2f}")

print("Yo'l haqi bo'yicha qiymatlar")
print(f"O'rtacha yo'l haqi: {uchinchi_klass['fare'].mean():.2f} $")
print(f"Minimal yo'l haqi: {uchinchi_klass['fare'].min():.2f} $")
print(f"Maximal yo'l haqi: {uchinchi_klass['fare'].max():.2f} $")
```

```python
# 1. Oilaviy o'lcham
df['family_size'] = df['sibsp'] + df['parch']

# 2. Jins bo'yicha ajratish
erkaklar_df = df[df['sex'] == 'male']
ayollar_df = df[df['sex'] == 'female']

# ==================== ERKAKLAR ====================
X_e = len(erkaklar_df)
Y_e = erkaklar_df['survived'].sum()
Z_e = (Y_e / X_e * 100) if X_e > 0 else 0

# Yoshni hisoblash (NaNlarni tashlab)
ages_e = erkaklar_df['age'].dropna()
A_e = ages_e.mean() if len(ages_e) > 0 else 0
A_min_e = ages_e.min() if len(ages_e) > 0 else 0
A_max_e = ages_e.max() if len(ages_e) > 0 else 0

B_e = erkaklar_df['fare'].mean() if len(erkaklar_df) > 0 else 0
C_e = erkaklar_df['family_size'].mean() if len(erkaklar_df) > 0 else 0

# Sinf bo'yicha taqsimot
class_counts_e = erkaklar_df['pclass'].value_counts().sort_index()
class_percents_e = (class_counts_e / X_e * 100) if X_e > 0 else [0,0,0]

print("Erkaklar:")
print(f"- Jami soni: {X_e}")
print(f"- Omon qolganlar: {Y_e} ({Z_e:.1f}%)")
print(f"- O'rtacha yosh: {A_e:.1f} (Min: {A_min_e:.1f}, Max: {A_max_e:.1f})")
print(f"- O'rtacha yo'l haqi: {B_e:.2f}$")
print(f"- O'rtacha oila a'zolari: {C_e:.1f}")
print("- Sinf bo'yicha taqsimoti:")
for cls in [1,2,3]:
    count = class_counts_e.get(cls,0)
    percent = class_percents_e.get(cls,0)
    print(f"  {cls}-sinf: {percent:.1f}% ({count} kishi)")
print("\n")

# ==================== AYOLLAR ====================
X_a = len(ayollar_df)
Y_a = ayollar_df['survived'].sum()
Z_a = (Y_a / X_a * 100) if X_a > 0 else 0

# Yoshni hisoblash (NaNlarni tashlab)
ages_a = ayollar_df['age'].dropna()
A_a = ages_a.mean() if len(ages_a) > 0 else 0
A_min_a = ages_a.min() if len(ages_a) > 0 else 0
A_max_a = ages_a.max() if len(ages_a) > 0 else 0

B_a = ayollar_df['fare'].mean() if len(ayollar_df) > 0 else 0
C_a = ayollar_df['family_size'].mean() if len(ayollar_df) > 0 else 0

# Sinf bo'yicha taqsimot
class_counts_a = ayollar_df['pclass'].value_counts().sort_index()
class_percents_a = (class_counts_a / X_a * 100) if X_a > 0 else [0,0,0]

print("Ayollar:")
print(f"- Jami soni: {X_a}")
print(f"- Omon qolganlar: {Y_a} ({Z_a:.1f}%)")
print(f"- O'rtacha yosh: {A_a:.1f} (Min: {A_min_a:.1f}, Max: {A_max_a:.1f})")
print(f"- O'rtacha yo'l haqi: {B_a:.2f}$")
print(f"- O'rtacha oila a'zolari: {C_a:.1f}")
print("- Sinf bo'yicha taqsimoti:")
for cls in [1,2,3]:
    count = class_counts_a.get(cls,0)
    percent = class_percents_a.get(cls,0)
    print(f"  {cls}-sinf: {percent:.1f}% ({count} kishi)")
```

```python
# 1. NaN qiymatlarni aniqlash va tozalash
nan_yosh_soni = df['age'].isnull().sum()
# Faqat yoshi bor yo'lovchilarni saqlab qolamiz
df_clean = df.dropna(subset=['age']).copy()

print(f"NAN qiymatlar soni = {nan_yosh_soni} ta")
print("-" * 30)

# 2. Yosh guruhlarini belgilash
def get_group(age):
    if age < 18: return 'Bolalar'
    elif 18 <= age <= 35: return 'Yoshlar'
    elif 36 <= age <= 60: return 'O\'rtacha yosh'
    else: return 'Keksalar'

df_clean['group'] = df_clean['age'].apply(get_group)

# 3. Guruhlar bo'yicha tahlil
guruhlar = ['Bolalar', 'Yoshlar', 'O\'rtacha yosh', 'Keksalar']

for g in guruhlar:
    subset = df_clean[df_clean['group'] == g]
    total = len(subset)
    survived = subset['survived'].sum()
    surv_pct = (survived / total * 100) if total > 0 else 0
    avg_fare = subset['fare'].mean()
    
    # Sinf taqsimoti
    pclass_total_pct = subset['pclass'].value_counts(normalize=True) * 100
    
    # Omon qolganlar sinf bo'yicha
    survived_class_count = subset[subset['survived'] == 1]['pclass'].value_counts()
    survived_class_pct = subset[subset['survived'] == 1]['pclass'].value_counts(normalize=True) * 100
    
    # Natijani formatlash: foiz (son)
    def format_survived(cls):
        pct = survived_class_pct.get(cls, 0)
        cnt = survived_class_count.get(cls, 0)
        return f"{pct:.1f}% ({cnt})"
    
    print(f"{g} ({' <18' if g=='Bolalar' else '18-35' if g=='Yoshlar' else '36-60' if g=='O\'rtacha yosh' else '60+' } yosh):")
    print(f"- Umumiy: {total} kishi")
    print(f"- Omon qolgan: {survived} kishi ({surv_pct:.1f}%)")
    print(f"- Sinf taqsimoti: 1-sinf: {pclass_total_pct.get(1, 0):.1f}%, 2-sinf: {pclass_total_pct.get(2, 0):.1f}%, 3-sinf: {pclass_total_pct.get(3, 0):.1f}%")
    print(f"- Omon qolganlar sinf bo'yicha: 1-sinf: {format_survived(1)}, 2-sinf: {format_survived(2)}, 3-sinf: {format_survived(3)}")
    print(f"- O'rtacha yo'l haqi: {avg_fare:.2f}$\n")
```
