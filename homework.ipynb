```python
import pandas as pd
import numpy as np
df = pd.read_csv('Titanic Dataset.csv')

# 1. NaN qiymatlarni aniqlash va tozalash
nan_yosh_soni = df['age'].isnull().sum()
# Faqat yoshi bor yo'lovchilarni saqlab qolamiz
df_clean = df.dropna(subset=['age']).copy()

print(f"NAN qiymatlar soni = {nan_yosh_soni} ta")
print("-" * 30)

# 2. Yosh guruhlarini belgilash
def get_group(age):
    if age < 18: return 'Bolalar'
    elif 18 <= age <= 35: return 'Yoshlar'
    elif 36 <= age <= 60: return 'O\'rtacha yosh'
    else: return 'Keksalar'

df_clean['group'] = df_clean['age'].apply(get_group)

# 3. Guruhlar bo'yicha tahlil
guruhlar = ['Bolalar', 'Yoshlar', 'O\'rtacha yosh', 'Keksalar']

for g in guruhlar:
    subset = df_clean[df_clean['group'] == g]
    total = len(subset)
    survived = subset['survived'].sum()
    surv_pct = (survived / total * 100) if total > 0 else 0
    avg_fare = subset['fare'].mean()
    
    # Sinf taqsimoti
    pclass = subset['pclass'].value_counts(normalize=True) * 100
    
    print(f"{g} ({' <18' if g=='Bolalar' else '18-35' if g=='Yoshlar' else '36-60' if g=='O\'rtacha yosh' else '60+' } yosh):")
    print(f"- Umumiy: {total} kishi")
    print(f"- Omon qolgan: {survived} kishi ({surv_pct:.1f}%)")
    print(f"- Sinf taqsimoti: 1-sinf: {pclass.get(1, 0):.1f}%, 2-sinf: {pclass.get(2, 0):.1f}%, 3-sinf: {pclass.get(3, 0):.1f}%")
    print(f"- O'rtacha yo'l haqi: {avg_fare:.2f}$\n")
```
