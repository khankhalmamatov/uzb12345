```python
import pandas as pd
import numpy as np

df = pd.read_csv('Titanic Dataset.csv')
df.info()
```

```python
df.describe()
```

```python
df_clean = df.drop(['name','ticket', 'cabin', 'embarked', 'boat', 'body', 'home.dest'], axis=1, inplace=True )
df_clean
# 1. Jami 
df['family_size'] = df['sibsp'] + df['parch']

# 2. sex
erkaklar_df = df[df['sex'] == 'male']
ayollar_df = df[df['sex'] == 'female']

# ERKAKLAR
X_e = len(erkaklar_df) # Jami soni
Y_e = erkaklar_df['survived'].sum() # Omon qolganlar soni
Z_e = (Y_e / X_e) * 100 # Omon qolish foizi
A_e = erkaklar_df['age'].mean() # O'rtacha yosh
B_e = erkaklar_df['fare'].mean() # O'rtacha yo'l haqi
C_e = erkaklar_df['family_size'].mean() # O'rtacha oila a'zolari

# Erkaklar sinfi taqsimoti:
P_e = (len(erkaklar_df[erkaklar_df['pclass'] == 1]) / X_e) * 100
Q_e = (len(erkaklar_df[erkaklar_df['pclass'] == 2]) / X_e) * 100
R_e = (len(erkaklar_df[erkaklar_df['pclass'] == 3]) / X_e) * 100

print(f"Erkaklar:")
print(f"- Jami soni: {X_e}")
print(f"- Omon qolganlar: {Y_e} ({Z_e:.1f}%)")
print(f"- O'rtacha yosh: {A_e:.1f}")
print(f"- O'rtacha yo'l haqi: {B_e:.2f}$")
print(f"- O'rtacha oila a'zolari: {C_e:.1f}")
print(f"- Sinf bo'yicha taqsimoti: 1-sinf: {P_e:.1f}%, 2-sinf: {Q_e:.1f}%, 3-sinf: {R_e:.1f}%  \n")



#  AYOLLAR 
X_a = len(ayollar_df) # Jami soni
Y_a = ayollar_df['survived'].sum() # Omon qolganlar soni
Z_a = (Y_a / X_a) * 100 # Omon qolish foizi
A_a = ayollar_df['age'].mean() # O'rtacha yosh
B_a = ayollar_df['fare'].mean() # O'rtacha yo'l haqi
C_a = ayollar_df['family_size'].mean() # O'rtacha oila a'zolari

# Ayollar sinfi taqsimoti:
P_a = (len(ayollar_df[ayollar_df['pclass'] == 1]) / X_a) * 100
Q_a = (len(ayollar_df[ayollar_df['pclass'] == 2]) / X_a) * 100
R_a = (len(ayollar_df[ayollar_df['pclass'] == 3]) / X_a) * 100

print(f"Ayollar:")
print(f"- Jami soni: {X_a} ")
print(f"- Omon qolganlar: {Y_a} ({Z_a:.1f}%)")
print(f"- O'rtacha yosh: {A_a:.1f}")
print(f"- O'rtacha yo'l haqi: {B_a:.2f}$")
print(f"- O'rtacha oila a'zolari: {C_a:.1f}")
print(f"- Sinf bo'yicha taqsimoti: 1-sinf: {P_a:.1f}%, 2-sinf: {Q_a:.1f}%, 3-sinf: {R_a:.1f}%")
```
